````python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
````

## **Scrapping Movies Data**

````python
# Specifying the URL from which movies related data will be fetched
url='https://www.justwatch.com/in/movies?release_year_from=2000'

# Sending an HTTP GET request to the URL
page=requests.get(url)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
soup=BeautifulSoup(page.text,'html.parser')
# Printing the prettified HTML content
#print(soup.prettify())
````

## **Fetching Movie URL's**

````python
# Write Your Code here
div=soup.find("div",class_="title-list-grid")    # finding all  div tag
atag=div.find_all("a")                          #finding all "a" tags
movie_link_list=[]                          #list to append the movie url
for i in atag:                              # using for loop to iterate among list of "a" tags
  http="https://www.justwatch.com"          #common link needs to be added after extracting the links from "href"
  temp_link=i.get("href")                   #extracting the link
  movie_link_list.append(http+temp_link)    #adding the common link + href links to get the complete url

movie_link_list_50=movie_link_list[:50]    #extracting 50 movie links
movie_link_list_50

````

## **Scrapping Movie Title**

````python
# Write Your Code here
import time            # importing time module to use time.sleep to have a delay before loading next link to avoid no text error
movie_title_list=[]                 #list to add the movies name
for i in movie_link_list_50:        #using for loop to iterate over movie link list
 url2=i
 page2=requests.get(url2)
 soup2=BeautifulSoup(page2.text,'html.parser')

 try:
  title_movie=soup2.find("div",class_="title-block")
  movie_title=title_movie.find("h1").text.strip()              #finding h1 with movie title , also using strip for additional removal of trailing spaces
  movie_title_list.append(movie_title)

 except Exception as err:                        #using exception to handle no text found error in movie title
    print(f'Error occurred: {err}')
    movie_title_list.append("Na")

 time.sleep(2)                                    #using sleep module to offer delay and to avoid over loading the server



print(movie_title_list)
print(len(movie_title_list))

````

## **Scrapping release Year**

````python
# Write Your Code here
year_list=[]
for i in movie_link_list_50:
 url2=i
 page2=requests.get(url2)
 soup2=BeautifulSoup(page2.text,'html.parser')

 try:
  year=soup2.find("span",class_="text-muted").text.strip()                #finding all span tags and using strip method to remove trailing spaces in year
  release_year= "".join([i for i in year if i not in ["(",")"]])          #using list comprehension and join method to remove paranthesis from year text
  year_list.append(release_year)                                           #appending to year _list

 except Exception as err:
  year_list.append("Na")
  print(f'Other error occurred: {err}')

 time.sleep(3)


print(year_list)
print(len(year_list))                                                        #checking length of list
````

## **Scrapping Genres**

````python
# Write Your Code here
genre_list=[]
for i in movie_link_list_50:
 url2=i
 page2=requests.get(url2)
 soup2=BeautifulSoup(page2.text,'html.parser')

 try :
  genre=soup2.find("h3",class_="detail-infos__subheading",string="Genres").find_next_sibling("div",class_="detail-infos__value").text.strip() #using find_next_sibling method to find the cosecutive tag under h3, i.e. div to extract the genre text
  genre_list.append(genre)
 except Exception as err:
  genre_list.append("Na")
  print(f'Other error occurred: {err}')
 time.sleep(3)


print(genre_list)
print(len(genre_list))
````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
IMDB_Rating=[]
for i in movie_link_list_50:
 url2=i
 page2=requests.get(url2)
 soup2=BeautifulSoup(page2.text,'html.parser')
 try:
  rating = soup2.find_all("div",class_="jw-scoring-listing__rating")[1].find("span").text.strip().split() #finding all div tag and using index 1 to scrape required text and spliting
  IMDB_Rating.append(rating[0])                                    #appending zeroth index only to  list
 except Exception as err:
  IMDB_Rating.append("Na")
  print(f'Other error occurred: {err}')
 time.sleep(3)

print(IMDB_Rating)
print(len(IMDB_Rating))
````

## **Scrapping Runtime/Duration**

````python
# Write Your Code here
runtime_list=[]
for i in movie_link_list_50:
 url2=i
 page2=requests.get(url2)
 soup2=BeautifulSoup(page2.text,'html.parser')

 try :
  runtime=soup2.find("h3",class_="detail-infos__subheading",string="Runtime").find_next_sibling("div",class_="detail-infos__value").text.strip()  #finfing all h3 tags to extract runtime
  runtime_list.append(runtime)
 except Exception as err:                     #handling error using try except block
  runtime_list.append("Na")
  print(f'Other error occurred: {err}')
 time.sleep(3)


print(runtime_list)
print(len(runtime_list))
````

## **Scrapping Age Rating**

````python
# Write Your Code here
age_rating=[]
for i in movie_link_list_50:
 url2=i
 page2=requests.get(url2)
 soup2=BeautifulSoup(page2.text,'html.parser')

 try :
  rating_age=soup2.find("h3",class_="detail-infos__subheading",string="Age rating").find_next_sibling("div",class_="detail-infos__value").text.strip() #extracting ageratings from h3 tags + div tag
  age_rating.append(rating_age)
 except Exception as err:
  print(f'Other error occurred: {err}')
  age_rating.append("Na")

 time.sleep(3)


print(age_rating)
print(len(age_rating))             #checking length of the list
````

## **Fetching Streaming Service Details**

````python
# Write Your Code here
streaming_service=[]
for i in movie_link_list_50:
 url2=i
 page2=requests.get(url2)
 soup2=BeautifulSoup(page2.text,'html.parser')

 try :
  streaming=soup2.find("div",class_="buybox-row stream").find_all("img")      #under div tag finding all imag tags
  temp_list=[]
  for i in range(len(streaming)):                                           #using for loop to iterate within the length of streaming (i.e -#number of streaming platforms)
    temp_list.append(streaming[i]["alt"])                               #appending the alt tag in temp_list
  temp_append=",".join(temp_list)                                        # after joining elements if more than 1 apending to

  streaming_service.append(temp_append)                                  #finally appending to main streaming_service list after converting to string
 except Exception as err:
  print(f'Other error occurred: {err}')
  streaming_service.append("Na")

 time.sleep(3)


print(streaming_service)
print(len(streaming_service))
````

## **Task 2 :- Data Filtering & Analysis**

````python
# Write Your Code here
#joining two data frames at axis =0
df=pd.concat([movie_df,tv_df],axis=0).reset_index()
df.drop('index',axis=1,inplace=True)
df

````

````python
df.info() # checking dtypes need to convert dates  and IMDB_Rating to numeric data type
````

````python
def check(n):         #function to check "Na" and replacing with Null values present in data frame or not
  if n == "Na":
    n=np.nan
  return n

df["IMDB_Rating"]=df["IMDB_Rating"].apply(check) #converting "Na" to Null to further convert into float

df["Available_Streaming_Services"]=df["Available_Streaming_Services"].apply(check) # removing and replacing Na values with null values

df["IMDB_Rating"]=df["IMDB_Rating"].astype("float") #changing data type of IMDB rating from object to numerical value

df["Release_Year"]=df["Release_Year"].astype("int")  #converting release year to int for comparison

df.info()
````

````python
#creating a new data frame with the filtered condition below
Present_Year=2024
condition1=df.loc[:,"Release_Year"]==Present_Year-1
condition2=df.loc[:,"Release_Year"]==Present_Year-2
condition3=df.loc[:,"IMDB_Rating"]>=7
Filtered_df = df.loc[(condition1|condition2)&(condition3)].reset_index().drop(["index"],axis=1)
Filtered_df
````
